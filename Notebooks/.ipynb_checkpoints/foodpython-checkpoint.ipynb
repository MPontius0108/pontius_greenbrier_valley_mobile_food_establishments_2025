```python
import matplotlib.pyplot as plt

occupation_list = ["Education", "Office/Admin", "Production Worker", "Office/Admin", "Executive/Management", "Sales", "Office/Admin", "Executive/Management", "Sales"]

# Replace spaces with underscores within each occupation
processed_list = [item.replace(" ", "_") for item in occupation_list]

# Join with spaces
text = " ".join(processed_list)

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("My Word Cloud")
plt.show()
```

```python
ingredients = pd.read_excel("C:/Users/user/workspace/c11/Capstone/Data/Recipe ingredients.xlsx")
ingredients
```

```python
# Corrected code
by_ingredient = ingredients['ingredient'].value_counts()
by_ingredient.head(60)
```

```python
# === 3. Fix column names (remove extra spaces) ===
ingredients.columns = ingredients.columns.str.strip()

# === 4. Convert amount per portion to float ===
ingredients['amount per portion'] = pd.to_numeric(ingredients['amount per portion'], errors='coerce')

# === 5. Group by ingredient: sum amount, keep most common unit ===
summary = (
    ingredients.groupby('ingredient')
    .agg(
        total_amount_per_portion=('amount per portion', 'sum'),
        portion_unit=('portion unit', lambda x: x.mode()[0] if not x.empty else 'unknown')
    )
    .reset_index()
    .sort_values('total_amount_per_portion', ascending=False)
)

# === 6. Display result ===
print(summary)
```
